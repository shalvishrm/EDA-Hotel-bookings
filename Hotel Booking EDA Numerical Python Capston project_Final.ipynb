## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# We implement the Emputation the dataset to clean the data

hotel_booking_data.copy()
## Handling missing /null values
## We do this opration column wise according to demand
````

````python
## Find the missing value, show the total null values for each column and sort it in descending order
hotel_booking_data.isnull().sum().sort_values(ascending=False)[:10]
````

````python
#Handling the missing/null values
````

````python
#create the new categories for both agent and company column
hotel_booking_data['agent']=hotel_booking_data['agent'].fillna('Not applicale')
hotel_booking_data['company'] = hotel_booking_data['company'].fillna('not applicable')
````

````python
#create the new categories for children and country columns
hotel_booking_data['children']=hotel_booking_data['children'].fillna('0')
hotel_booking_data['country']=hotel_booking_data['country'].fillna('Unknown')
````

````python
#check null value again
hotel_booking_data.isnull().sum()
````

````python
# Change Value 1 and 0 with Yes Or No
# 0 ---> No
# 1 ---> Yes
hotel_booking_data.loc[hotel_booking_data["is_repeated_guest"] == 1, "is_repeated_guest"] = "Yes"
hotel_booking_data.loc[hotel_booking_data["is_repeated_guest"] == 0, "is_repeated_guest"] = "No"
````

````python
# Change Value 1 and 0 with Yes Or No
# 0 ---> No
# 1 ---> Yes
hotel_booking_data.loc[hotel_booking_data["required_car_parking_spaces"]==1,"required_car_parking_spaces"]="Yes"
hotel_booking_data.loc[hotel_booking_data["required_car_parking_spaces"]==0,"required_car_parking_spaces"]="No"
````
